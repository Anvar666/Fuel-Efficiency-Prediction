# Description of each column

mpg: Miles per gallon, representing the fuel efficiency of the car in terms of how many miles it can travel per gallon of fuel consumed. Higher values indicate better fuel efficiency.

cylinders: The number of cylinders in the engine. This typically ranges from 3 to 12 cylinders, with more cylinders generally indicating a more powerful engine.

displacement: The total volume of all cylinders in the engine in cubic inches or liters. It represents the engine's capacity, with larger values generally indicating more powerful engines.

horsepower: The power output of the engine, usually measured in horsepower. Higher horsepower values generally indicate higher engine performance.

weight: The weight of the car, typically measured in pounds or kilograms. Heavier cars tend to consume more fuel.

acceleration: The time it takes for the car to accelerate from 0 to 60 miles per hour (or 0 to 100 kilometers per hour), usually measured in seconds. Lower values indicate quicker acceleration.

model year: The year when the car model was manufactured. It can be used to track improvements in fuel efficiency and technological advancements over time.

origin: The country of origin or manufacture of the car. This could be represented numerically or categorically, with values such as 1 for American, 2 for European, and 3 for Asian, for example.

car name: The name or identifier of the car model.

# Encoding

# Prediction - Regression

```python
x=a.iloc[:,[1,2,3,4,5,6,7]].values                                   # Independent values
x
```

```python
y=a.iloc[:,0].values                                          # Dependent value
y
```

# Train, Test Splitting

```python
from sklearn.model_selection import train_test_split                       
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=80,random_state=35)
```

```python
x_train
```

```python
x_test
```

# Linear

```python
from sklearn.linear_model import LinearRegression                        # Importing Linear Regression
```

```python
regressor=LinearRegression()
regressor
```

```python
regressor.fit(x_train,y_train)                                          # Fitting on Regressor
```

```python
xpred1=regressor.predict(x_train)
xpred1
```

```python
ypred1=regressor.predict(x_test)
ypred1
```

```python
trainingscore1=regressor.score(x_train,y_train)
trainingscore1
```

```python
testingscore1=regressor.score(x_test,y_test)
testingscore1
```

```python
from sklearn import metrics
```

```python
print("mae"),metrics.mean_absolute_error(y_test,ypred1)
```

```python
print("mse"),metrics.mean_squared_error(y_test,ypred1)
```

```python
print("rmse"),np.sqrt(metrics.mean_squared_error(y_test,ypred1))
```

```python
print("rmae"),np.sqrt(metrics.mean_absolute_error(y_test,ypred1))
```

```python
from sklearn.metrics import r2_score
```

```python
r1=r2_score(y_test,ypred1)
r1
```
